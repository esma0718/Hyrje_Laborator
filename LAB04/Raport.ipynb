# Shembull: Fit linear

```python
import numpy as np

x = np.array([1,2,3,4])
y = np.array([2,4,6,8])

coeff = np.polyfit(x,y,1)

slope = coeff[0]
intercept = coeff[1]
```

```python
y = slope * x + intercept
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4])
y = np.array([2.1,4.2,6.1,8.0])

coeff = np.polyfit(x,y,1)

fit = coeff[0]*x + coeff[1]

plt.scatter(x,y,label="Data")
plt.plot(x,fit,label="Fit")

plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.show()
```
