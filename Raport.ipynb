````python
# Importet standarde të kursit
import numpy as np
import matplotlib.pyplot as plt

np.__version__
````

````python
# Test minimal i NumPy
data = np.array([10.1, 10.2, 9.9, 10.0, 10.3])
mean = np.mean(data)
std_sample = np.std(data, ddof=1)

mean, std_sample
````

````python
# Test minimal i Matplotlib
plt.hist(data, bins=5)
plt.xlabel("Vlera (njësi)")
plt.ylabel("Frekuenca")
plt.title("Histogram test")
plt.grid(True)
plt.show()
````
